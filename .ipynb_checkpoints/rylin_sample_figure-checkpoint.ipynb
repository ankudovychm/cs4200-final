```python
import pandas as pd
import altair as alt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
data = pd.read_csv('data.csv')

data.head()
```

```python
sample = data.sample(n=2000, random_state=3) 
```

```python
speech = alt.Chart(sample).mark_point().encode(
    alt.Y("popularity:Q", title = 'Popularity'),
    alt.X("speechiness:Q", title = 'Speechiness')
)

dance = alt.Chart(sample).mark_point().encode(
    alt.Y("popularity:Q", title = 'Popularity'),
    alt.X("danceability:Q", title = 'Danceability')
)

acoustic = alt.Chart(sample).mark_point().encode(
    alt.Y("popularity:Q", title = 'Popularity'),
    alt.X("acousticness:Q", title = 'Acousticness')
)

loud = alt.Chart(sample).mark_point().encode(
    alt.Y("popularity:Q", title = 'Popularity'),
    alt.X("loudness:Q", title = 'Loudness')
)

live = alt.Chart(sample).mark_point().encode(
    alt.Y("popularity:Q", title = 'Popularity'),
    alt.X("liveness:Q", title = 'Liveness')
)

instrument = alt.Chart(sample).mark_point().encode(
    alt.Y("popularity:Q", title = 'Popularity'),
    alt.X("instrumentalness:Q", title = 'Instrumentalness')
)

text = alt.Chart().mark_text(
    fontSize=15,
    fontWeight=600,
    dx=500,
    dy=0,
).encode(
    text=alt.value("Song Popularity by Musical Qualities")
)

text&(speech&dance|acoustic&loud|live&instrument)
```
